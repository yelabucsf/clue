```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
original = pd.read_csv('/data/codec/production.run/exp.csvs/original.csv')
prestim = pd.read_csv('/data/codec/production.run/exp.csvs/prestim.csv')
poststim = pd.read_csv('/data/codec/production.run/exp.csvs/poststim.csv')
postpool = pd.read_csv('/data/codec/production.run/exp.csvs/post.pool.counts.csv')
```

```python
sns.stripplot(data=original,x='source',y='viability');
plt.ylim(0, 100);
```

```python
sns.stripplot(data=prestim,x='source',y='viability');
plt.ylim(0, 100);
```

```python
sns.stripplot(data=prestim,x='source',y='cell_number');
```

```python
plt.figure(figsize=(25,5))
sns.barplot(data=poststim,x='well',y='viability');
plt.ylim(0, 100);
```

```python
v, c = np.unique(poststim['exp_id'].values,return_counts=True)
poststim_repeats = poststim[poststim['exp_id'].isin(v[c >1])]
```

```python
sns.stripplot(data=poststim_repeats, x='condition',y='cell_number');
sns.boxplot(data=poststim_repeats, x='condition',y='cell_number', boxprops=dict(alpha=.3), showfliers=False);
plt.ylim(0, 700000);
```

```python
sns.stripplot(data=poststim_repeats, x='condition',y='viability');
sns.boxplot(data=poststim_repeats, x='condition',y='viability', boxprops=dict(alpha=.3), showfliers=False);
plt.ylim(0, 100);
```

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=poststim_repeats, x='exp_id',y='cell_number', jitter=False);
plt.ylim(0, 700000);
```

```python
sns.stripplot(data=postpool, x='pool',y='viability', hue='count_type');
plt.ylim(0, 100);
```
